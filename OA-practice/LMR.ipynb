### Best execution - Volume Weighted Average Price

```python
df['quantity'] = df[['TradeValueUSD', 'PriceUSD']].apply(lambda x: x[0]//x[1], axis=1)
df['count'] = df.groupby('TradeValueUSD')['TradeValueUSD'].transform('count')
```

```python
df['cum_vol'] = df['quantity'].cumsum()
df['cum_vol_price'] = (df['quantity'] * df['PriceUSD']).cumsum()
df['VWAP'] = df['cum_vol_price'] / df['cum_vol']
df['Faraway %'] = df[['PriceUSD', "VWAP"]].apply(lambda x: abs((x[1]-x[0])/x[0]), axis=1)
```

```python
df.sort_values(by=['Faraway %'], ascending=False)[:10]
```
